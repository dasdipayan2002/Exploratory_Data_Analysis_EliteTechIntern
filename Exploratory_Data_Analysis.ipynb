```python
# Importing required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading the diamonds dataset fro seaborn
diamonds = sns.load_dataset('diamonds')
```

```python
diamonds.head()
```

```python
# Displaying the shape and basic information of the dataset 
print(f"Dataset Shape : {diamonds.shape}")
diamonds.info()
```

```python
# Displaying the statistical Summary of the dataset
diamonds.describe()
```

```python
# Checking for missing values
missing_values = diamonds.isnull().sum()
print("Missing Values in each column : \n",missing_values)
```

```python
# plotting the distribution of diamond prices
plt.figure(figsize=(10, 5))
sns.histplot(diamonds['price'], bins=50, kde=True)
plt.title("Distribution of Diamond Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# Box plot of price vs cut
plt.figure(figsize=(10, 5))
sns.boxplot(x='cut',y='price',data=diamonds)
plt.title("Diamond price by Cut")
plt.show()
```

```python
# Box plot of price vs color
plt.figure(figsize=(10, 5))
sns.boxplot(x='color',y='price',data=diamonds)
plt.title("Diamond price by Color")
plt.show()
```

```python
# scatter plot of carat vs price
plt.figure(figsize=(10, 5))
sns.scatterplot(x='carat',y='price',data=diamonds)
plt.title("Carat vs Price")
plt.xlabel("Carat")
plt.ylabel("Price")
plt.show()
```

```python
# pair plot for numerical features
sns.pairplot(diamonds[['carat','depth','table','price']])
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(diamonds.corr(numeric_only=True),annot=True,cmap='coolwarm',vmin=-1,vmax=1)
plt.title("Correlation Matrix")
plt.show()
```
